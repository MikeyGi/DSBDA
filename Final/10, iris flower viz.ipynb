```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
%matplotlib inline
```

```python
fig , axes = plt.subplots(2,2 , figsize=(20,10))

axes[0,0].hist(df['sepal_length'])
axes[0,0].set_title("SEPAL LENGTH")

axes[0,1].hist(df['sepal_width'])
axes[0,1].set_title("SEPAL LENGTH")

axes[1,0].hist(df['petal_length'])
axes[0,1].set_title("SEPAL LENGTH")

axes[1,1].hist(df['petal_width'])
axes[0,1].set_title("SEPAL LENGTH")
```

```python
sns.boxplot(x='species',y='sepal_length' , data=df ,palette='icefire')
```
